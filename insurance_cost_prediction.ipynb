# **Removing Outliers**

```python
Q1 = df['Weight'].quantile(0.25)
Q3 = df['Weight'].quantile(0.75)
IQR = Q3 - Q1
threshold = 1.5
outliers = df[(df['Weight'] < Q1 - threshold * IQR) | (df['Weight'] > Q3 + threshold * IQR)]
```

```python
outliers.shape
```

```python
Q2 = df['PremiumPrice'].quantile(0.25)
Q4 = df['PremiumPrice'].quantile(0.75)
IQR2 = Q4 - Q2

# identify outliers
threshold2 = 1.5
outliers2 = df[(df['PremiumPrice'] < Q2 - threshold2 * IQR2) | (df['PremiumPrice'] > Q4 + threshold2 * IQR2)]
```

```python
outliers2.shape
```

```python
df1=df[~df.index.isin(outliers.index)]
```

```python
df2=df1[~df1.index.isin(outliers2.index)]
```

```python
df2.shape
```

# **Hypothesis Testing**

```python
from scipy import stats
```

Testing whether or not the presence of Diabetes leads to an increase in the Premium Price. We will use the two sample T-test to compare the means of the samples

Null Hypothesis : Diabetes does not lead to an increase in the premium price

Alternative Hypothesis : Diabetes leads to an increase in the premium price

Threshold p-value=5%

```python
stats.ttest_ind(df2[df2['Diabetes']==1]['PremiumPrice'], df2[df2['Diabetes']==0]['PremiumPrice'], alternative='greater')
```

**We reject the null hypothesis. The test shows a strong tendency for the Premium Price to increase in the presence of diabetes**

Testing whether or not the presence of known allergies leads to an increase in the Premium Price. We will use the two sample T-test to compare the means of the samples

Null Hypothesis : The presence of known allergies does not lead to an increase in the premium price

Alternative Hypothesis : The presence of known allergies leads to an increase in the premium price

Threshold p-value=5%

```python
stats.ttest_ind(df2[df2['KnownAllergies']==1]['PremiumPrice'], df2[df2['KnownAllergies']==0]['PremiumPrice'], alternative='greater')
```


**We fail to reject the null hypothesis. There isn't sufficient evidence that the presence of known allergies leads to an increase in premium price**

Testing whether or not the presence of transplants leads to an increase in the Premium Price. We will use the two sample T-test to compare the means of the samples

Null Hypothesis : Presence of transplants does not lead to an increase in the premium price

Alternative Hypothesis : transplants leads to an increase in the premium price

Threshold p-value=5%

```python
stats.ttest_ind(df2[df2['AnyTransplants']==1]['PremiumPrice'], df2[df2['AnyTransplants']==0]['PremiumPrice'], alternative='greater')
```


**We reject the null hypothesis. There is sufficient evidence that the presence of transplants leads to an increase in premium price**

Testing whether or not blood pressure problmes lead to an increase in the Premium Price. We will use the two sample T-test to compare the means of the samples

Null Hypothesis : Blood pressure problems do not lead to an increase in the premium price

Alternative Hypothesis : Blood pressure problems lead to an increase in the premium price

Threshold p-value=5%

```python
stats.ttest_ind(df2[df2['BloodPressureProblems']==1]['PremiumPrice'], df2[df2['BloodPressureProblems']==0]['PremiumPrice'], alternative='greater')
```


**We reject the null hypothesis. There is sufficient evidence that the presence of blood pressure problems leads to an increase in premium price**


Testing whether or not a history of cancer in family lead to an increase in the Premium Price. We will use the two sample T-test to compare the means of the samples

Alternative Hypothesis : The presence of chronic diseases increase in the premium price

Testing whether or not the number of major surgeries leads to a change in the Premium Price. We will use the ANOVA test to compare the means of the samples

Null Hypothesis : The number of major surgeries has no effec on the premium price

Alternative Hypothesis : The number of major surgeries makes a difference to the premium price

Threshold p-value=5%

```python
stats.f_oneway(df2[df2['NumberOfMajorSurgeries']==0]['PremiumPrice'], df2[df2['NumberOfMajorSurgeries']==1]['PremiumPrice'], df2[df2['NumberOfMajorSurgeries']==2]['PremiumPrice'], df2[df2['NumberOfMajorSurgeries']==3]['PremiumPrice'])
```


**We reject the null hypothesis. There is sufficient evidence that the number of major surgeries has an effect on the premium price**



Testing whether or not the number of major surgeries leads to an increases in the Premium Price. We will use the Pearson rank correlaton coefficient test to compare the data

Null Hypothesis : The number of major surgeries leads to no increase in the premium price

Alternative Hypothesis : The number of major surgeries leads to an increase in the premium price

Threshold p-value=5%

```python
stats.pearsonr(df2['PremiumPrice'], df2['NumberOfMajorSurgeries'], alternative='greater')
```


**We reject the null hypothesis. There is sufficient evidence that the number of major surgeries leads to an increase in the premium price**



# **Data Preprocessing**

```python
df=df2
```

```python
df.isnull().sum()
```

There are no missing values in the dataset

# **Feature Engineering**

```python
df['bmi']=df['Weight']/((df['Height']/100)**2)
```

```python
df['obese']=(df['bmi']>30)
df['obese']=df['obese'].astype('int')
df['healthyweight']=(df['bmi']>18.5)*(df['bmi']<24.9)
df['healthyweight']=df['healthyweight'].astype('int')
df['underweight']=(df['bmi']<18.5)
df['underweight']=df['underweight'].astype('int')
df['overweight']=(df['bmi']>24.9)*(df['bmi']<30)
df['overweight']=df['overweight'].astype('int')
```

**Encoding the presence of two or more risky factors in one person using a feature called 'RiskyPerson'**

```python
df['riskyperson']=df['obese']*(df['NumberOfMajorSurgeries']>0)+df['obese']*df['AnyTransplants']+df['obese']*df['Diabetes']
+df['obese']*df['AnyChronicDiseases']+df['obese']*df['BloodPressureProblems']+(df['NumberOfMajorSurgeries']>0)*df['AnyTransplants']+(df['NumberOfMajorSurgeries']>0)*df['Diabetes']+(df['NumberOfMajorSurgeries']>0)*df['AnyChronicDiseases']+(df['NumberOfMajorSurgeries']>0)*df['BloodPressureProblems']+df['AnyTransplants']*df['Diabetes']+df['AnyTransplants']*df['AnyChronicDiseases']*df['BloodPressureProblems']+df['Diabetes']*df['AnyChronicDiseases']+df['Diabetes']*df['BloodPressureProblems']
+df['AnyChronicDiseases']*df['BloodPressureProblems']
```

# **Training the data**

```python
y=df['PremiumPrice']
```

```python
x=df.drop(['PremiumPrice'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.metrics import mean_absolute_percentage_error, r2_score
```

**Gradient Boosting Regressor Model**

```python
from sklearn.ensemble import GradientBoostingRegressor
reg=GradientBoostingRegressor()
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = {
'n_estimators' : [250, 500, 750,1000,1250],
'max_depth' : [10,20,30,40,50],
'min_samples_split' : [2, 4,6,8,10],
'min_samples_leaf' : [1, 3,5,7,9],
}
```

```python
clf = RandomizedSearchCV(reg, params, verbose=10,n_jobs=-1,scoring=r2_score)
clf.fit(x_train, y_train) #takes 5 minutes to run
```

```python
clf.best_params_
```

```python
r2_score(y_test,clf.predict(x_test))
```

```python
mean_absolute_percentage_error(y_test, clf.predict(x_test))
```

```python
best_predictor=best_clf = clf.best_estimator_
```

The Gradient Boosting Regressor has a better MAPE and R2- score compared to the Linear Regressor

**Random Forest Model**

```python
from sklearn.ensemble import RandomForestRegressor
reg2=RandomForestRegressor()
params2 = {
'n_estimators' : [250, 500, 750,1000,1250],
'max_depth' : [10,20,30,40,50],
'min_samples_split' : [2, 4,6,8,10],
'min_samples_leaf' : [1, 3,5,7,9],
}

clf2 = RandomizedSearchCV(reg2, params2, verbose=10,n_jobs=-1,scoring=r2_score)
clf2.fit(x_train, y_train) # takes 5 minutes to run

```

```python
clf2.best_params_
```

```python
r2_score(y_test, clf2.predict(x_test))
```

```python
mean_absolute_percentage_error(y_test, clf2.predict(x_test))
```

```python
best_predictor2=clf2.best_estimator_
```

The Random Forest Model has a better MAPE and R2 compared to the Gradient Boosting Regressor
